```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import caffe
import os, sys
from funcs import classifier

PROJECT_ROOT = "/Users/vikeshkhanna/stanford/cs231n/project"
TRAIN_ROOT = os.path.join(PROJECT_ROOT, "train")
SNAPSHOT_ROOT = os.path.join(TRAIN_ROOT, "snapshots")
DATA_ROOT = os.path.join(PROJECT_ROOT, "data")

print "CAFFE_ROOT=%s" % PROJECT_ROOT
print "TRAIN_ROOT=%s" % TRAIN_ROOT
print "SNAPSHOT_ROOT=%s" % SNAPSHOT_ROOT

RECOGNIZER_MODEL_FILE = os.path.join(os.path.join(TRAIN_ROOT, "mnist"), "char74k_recognizer_deploy.prototxt")
RECOGNIZER_PRETRAINED = os.path.join(SNAPSHOT_ROOT, "recognizer_iter_15000.caffemodel")

DETECTOR_MODEL_FILE = os.path.join(os.path.join(TRAIN_ROOT, "mnist"), "char74k_detector_deploy.prototxt")
DETECTOR_PRETRAINED = os.path.join(SNAPSHOT_ROOT, "detector_iter_15000.caffemodel")

TRAIN_XML = os.path.join(os.path.join(DATA_ROOT), "svt/train.xml")
TEST_XML = os.path.join(os.path.join(DATA_ROOT), "svt/test.xml")

print "RECOGNIZER_MODEL_FILE=%s" % RECOGNIZER_MODEL_FILE
print "RECOGNIZER_PRETRAINED=%s" % RECOGNIZER_PRETRAINED
```

```python
# Load XML files for the SVT dataset
from bs4 import BeautifulSoup

svt_train_file = open(TRAIN_XML)
svt_test_file = open(TEST_XML)

svt_train = BeautifulSoup(svt_train_file.read())
svt_test = BeautifulSoup(svt_test_file.read())

svt_train_file.close()
svt_test_file.close()
```

```python
# Create recognizer and detector CNNs
recognizer = classifier.Recognizer(RECOGNIZER_MODEL_FILE, RECOGNIZER_PRETRAINED)
detector = classifier.Detector(DETECTOR_MODEL_FILE, DETECTOR_PRETRAINED)
```

```python
IMAGE_FILE = os.path.join(DATA_ROOT, "English/Img/GoodImg/Bmp/Sample036/img036-00002.png")
print "IMAGE_FILE=%s" % IMAGE_FILE
input_image = caffe.io.load_image(IMAGE_FILE)
plt.imshow(input_image)
```

```python
prediction = recognizer.recognize([input_image])
print 'Prediction Shape: ', prediction[0].shape
plt.plot(prediction[0])
print 'Prediction class: ', classifier.transform_label(prediction[0].argmax())
```

```python
prediction = detector.detect([input_image])
print 'Prediction Shape: ', prediction[0].shape
plt.plot(prediction[0])
print 'Prediction class: ', classifier.transform_label(prediction[0].argmax())
```
