```python
# Vectorized
# Fraction of current size

masks = [(60, 60)]
stride = 5
# masks = [16, 8, 4, 2]
TEST_IMG = "/Users/vikeshkhanna/Downloads/word/word/1/10.jpg"
# TEST_IMG = os.path.join(SVT_ROOT, "img/00_06.jpg")
TEST_IMG = "/Users/vikeshkhanna/Downloads/test3.jpg"

original_img = caffe.io.load_image (TEST_IMG)
# Padding

for scale1, scale2 in masks:
    img = np.lib.pad(original_img, ((scale1/4,scale1/4), (scale2/4,scale2/4), (0,0)), 'edge')
    mask = (scale1, scale2, 3)
    plt.title("Scale: %d, %d; Shape=%s" % (scale1, scale2, img.shape))
    plt.imshow(img)
    plt.show()
    
    patches = image.extract_patches (img, mask, stride)[:,:,0,:,:]
    print patches.shape
    
    patches_flat = np.reshape(patches, (patches.shape[0]*patches.shape[1], mask[0], mask[1], mask[2]))
    plt.title ("Random patch. Shape=%s, flat_shape:%s" % (str(patches.shape), str(patches_flat.shape)))
    plt.imshow(patches_flat[-1])
    plt.show()
    
    print ("Detecting patches...")
    sys.stdout.flush()
    predictions_detect = detector.detect(patches_flat)
    predictions_detect_img = np.reshape(predictions_detect[:,1], (patches.shape[0], patches.shape[1]))
    plt.imshow(predictions_detect_img)
    plt.show()
    
    print ("Recognizing patches...")
    sys.stdout.flush()
    predictions_recognize = recognizer.recognize(patches_flat)
    predictions_recognize_img = np.reshape(np.max(predictions_recognize, axis=1), (patches.shape[0], patches.shape[1]))
    
    predictions_recognize_label = np.argmax(predictions_recognize, axis=1)
    f = np.vectorize(lambda x:classifier.transform_label(x))
    predictions_recognize_transformed_label = f(predictions_recognize_label)
    predictions_recognize_transformed_label_img = np.reshape(predictions_recognize_transformed_label, 
                                                            (patches.shape[0], patches.shape[1]))
    plt.imshow(predictions_recognize_img)
    plt.show()
```

```python
prob_map = predictions_detect_img
# prob_map[prob_map < 0.9] = 0

threshold = 0.90
indices = np.where(prob_map > threshold)
plt.imshow(prob_map > threshold)
plt.show()

for i, r in enumerate(indices[0]):
    c = indices[1][i]
    original_r = r*stride
    original_c = c*stride
    plt.title("Char: %s, Prob: %f" % (predictions_recognize_transformed_label_img[r,c], prob_map[r,c]))
    plt.imshow(img[original_r:original_r+scale1, original_c:original_c+scale2])
    plt.show()
```

```python
from scipy import signal

best_row = None
best_prob = -1

for row in range(prob_map.shape[0]):
    print "Row : %d" % row
    probs = prob_map[row,:]
    peakind = signal.find_peaks_cwt(probs, np.arange(1,10))
    print peakind
    
    plt.plot(probs)
    plt.show()
    sum_prob = 0
    
    av_peak_distance = np.mean(np.abs(np.ediff1d(peakind)))
    print ("Average peak distance: %f" % av_peak_distance)
    print "Variance of distances from av_peak_distance: ", np.abs(np.abs(np.ediff1d(peakind)) - av_peak_distance)
    
    for column in peakind:
        sum_prob += prob_map[row, column]
        plt.title("Char: %s, Prob:%f, Column:%d" %(predictions_recognize_transformed_label_img[row, column], 
                                                   predictions_recognize_img[row, column],
                                                   column))
        plt.imshow(img[row*stride:row*stride+scale1, column*stride:column*stride+scale2])
        plt.show()
    
    av_prob = sum_prob / len(peakind)
    
    if av_prob > best_prob:
        best_prob = av_prob
        best_row = row
        
print best_row, best_prob
```
