```python
from scipy import signal
from scipy.ndimage.filters import maximum_filter
import pprint

best_row = None
best_prob = -1
stats = {}

for row in range(predictions_detect_img.shape[0]):
    probs = predictions_detect_img[row,:]
    mean_prob = np.mean(probs)
    
    # peakind = signal.find_peaks_cwt(probs, np.arange(1,10), min_snr=10)
    peakind = np.nonzero(maximum_filter (probs, size=(scale2/stride)*0.75) == probs)[0]
    mean_best = np.mean(probs[peakind])
    
    # Remove everything not within 0.7 of max
    peakind = peakind[probs[peakind] > 0.5*np.max(probs)]
    
    if mean_best < 0.5:
        continue
    
    # print "Row : %d" % row
    # print peakind
    
    stats[row] = {}
    
    x = np.array([])
    y = np.array([])
    
    for i, prob in enumerate(probs):
        x = np.append(x, scale2/2 + i*stride)
        y = np.append(y, prob)
        
    # plt.plot(x,y)
    # plt.show()
    
    sum_prob_det = 0
    sum_prob_rec = 0
    sum_prob_combined = 1
    sum_rec_conf = 0
    
    av_peak_distance = np.mean(np.abs(np.ediff1d(peakind)))
    av_peak_ind_distance = np.abs(np.abs(np.ediff1d(peakind)) - av_peak_distance)
    av_peak_ind_distance_mean = np.mean(av_peak_ind_distance)
    stats[row][AVG_PEAK_DISTANCE] = av_peak_distance
    stats[row][VAR_PEAK_DISTANCE] = av_peak_ind_distance
    stats[row][MEAN_PEAK_VARIANCE] =  av_peak_ind_distance_mean 
    word = []
    
    for column in peakind:
        # Ugh
        preds_rec = np.reshape(predictions_recognize, (patches.shape[0], patches.shape[1], -1))[row, column] 
        rec_conf = np.abs(np.ediff1d(preds_rec[np.argsort(preds_rec)[-2:]])[0])
        
        label = predictions_recognize_transformed_label_img[row, column]
        prob_rec = predictions_recognize_img[row, column]
        prob_det = predictions_detect_img[row, column]
        prob_combined = prob_det*prob_rec
        word.append(label)
        
        sum_prob_det += prob_det
        sum_prob_rec += prob_rec
        sum_prob_combined *= prob_combined
        sum_rec_conf += rec_conf
        '''
        plt.title("Char: %s (%d), Recog_Prob:%f, Detection_Prob: %f, Combined_Prob: %f, Rec Confidence: %f, Column:%d" 
                                                  %(label,
                                                   predictions_recognize_label_img[row, column],
                                                   prob_rec,
                                                   prob_det,
                                                   prob_combined,
                                                   rec_conf,
                                                   column))
        plt.imshow(img[row*stride:row*stride+scale1, column*stride:column*stride+scale2])
        plt.show()
        '''
        
    if len(peakind) > 0:
        av_prob = sum_prob_det / len(peakind)
        score = av_prob 
        
        if score > best_prob :
            best_prob = score
            best_row = row
        
        stats[row]["Average probabiliy of detection"] = av_prob
        stats[row]["Average probability of recognition"] = sum_prob_rec / len(peakind)
        stats[row]["Average probability of combined"] = sum_prob_combined / len(peakind)
        stats[row]["Average rec conf"] = sum_rec_conf / len(peakind)
        stats[row]["Word"] = "".join(word)
        stats[row][PEAKIND] = peakind
        stats[row]["Score"] = score
    # print "Row Stats..."
    # pprint.pprint(stats[row])

print "Best row: ", best_row
pprint.pprint(stats[best_row])
```

```python
import itertools

# Fuzzy search for best confidence in the detector response
# Returns top k chars for each peak in detector response.
def fuzzy_search(bbImg, mask, stride, k):
    # Fuzzy search close to which detector peak. Start with first.
    scale1, scale2, channel = mask
    peak_align_index = 0
    rpatches = image.extract_patches (bbImg, mask, stride)[0,:,0,:,:]
    rpatches_recognize = recognizer.recognize(rpatches)
    rpatches_detector = detector.detect(rpatches)
    
    probs = rpatches_detector[:,1]
    peakind = np.nonzero(maximum_filter (probs, size=(scale2/stride)*0.75) == probs)[0]
    peakind = peakind[probs[peakind] > 0.7*np.max(probs)]
    plt.plot(probs)
    plt.show()
    
    print "Permute by: ", len(peakind)
    # Dictionary mapping the alignment index in peakind with all the 
    # recognitions (and their stats). We'll take the chars with top k
    # confidence margins
    align = {}
    
    # List of recognitions aligned to each detector resposne. We'll choose
    # top k.
    for i, col in enumerate(peakind):
        align[i] = []
    
    for col, pred_recognize in enumerate(rpatches_recognize):
        label = np.argmax(pred_recognize)
        char = classifier.transform_label (label)
        rec_conf = np.abs(np.ediff1d(pred_recognize[pred_recognize.argsort()[-2:]])[0])
        prob_recognizer = np.max(pred_recognize)
        prob_detector = rpatches_detector[col][1]
        rstat = { DETECTION_PROB: prob_detector, RECOGNITION_PROB: prob_recognizer, 
                 RECOGNITION_CONFIDENCE: rec_conf, CHAR_LABEL : char, POSITION: col}
        
        if peak_align_index < len(peakind) - 1:
            # Next one is closer
            if abs(peakind[peak_align_index + 1] - col) < abs(peakind[peak_align_index] - col):
                peak_align_index += 1
        
        align[peak_align_index].append(rstat)
    
    # Choose top k for each index
    results = [[] for i in range(len(peakind))]
    
    for i, rstats_list in align.iteritems():
        print ("Row: %d" % i)
        best_rstats_list = sorted(rstats_list, key = lambda x: x[RECOGNITION_CONFIDENCE], 
                                    reverse=True)[:k]
        
        # Augment align with a blank character for a peak with low detector response
        if rpatches_detector[peakind[i]][1] < 0.9:
            blank_dict = { DETECTION_PROB: 1, RECOGNITION_PROB: 1, 
                 RECOGNITION_CONFIDENCE: 1, CHAR_LABEL : "", POSITION: peakind[i]}
            best_rstats_list.append(blank_dict)
            print "%d is fuzzy" % i
        results[i].extend(best_rstats_list)
        pprint.pprint(results[i])
    
    candidates = []
    
    for combination in itertools.product(*results):
        chars = []
        av_rec_conf = 0
        av_det_prob = 0
        av_rec_prob = 0
        cnt = 0
        
        # Careful. There might be a blank.
        for rstat in combination:
            char = rstat[CHAR_LABEL]
            
            if char != "":
                chars.append(rstat[CHAR_LABEL])
                av_rec_conf += rstat[RECOGNITION_CONFIDENCE]
                av_det_prob += rstat[DETECTION_PROB]
                av_rec_prob += rstat[RECOGNITION_PROB]
                cnt += 1
                
        word = "".join(chars)
        candidate_dict = { WORD: word, RECOGNITION_CONFIDENCE: av_rec_conf / cnt,
                          RECOGNITION_PROB: av_rec_prob / cnt,
                          DETECTION_PROB: av_det_prob / cnt
                          }
        candidates.append(candidate_dict)
    
    return candidates

peakind = stats[best_row]["peakind"]
h_low, h_high = max(stride*best_row-stride, 0), stride*best_row + scale1*1.1
w_low, w_high = max(stride*peakind[0]-stride, 0), stride*peakind[-1] + scale2*1.1
bbImg = img[h_low:h_high, w_low:w_high,:]
plt.title("BbImg")
plt.imshow(bbImg)
plt.show()

# print predictions_detect_img[best_row, peakind]
pprint.pprint(stats[best_row])
plt.plot(predictions_detect_img[best_row,:])
plt.show()

print ("Finding candidates...")
sys.stdout.flush()
candidates = fuzzy_search(bbImg, mask, stride, 3)
print ("Candidates ready...")
sorted_candidates = sorted(candidates, key=lambda x: x[RECOGNITION_CONFIDENCE] + unigram[x[WORD].lower()], reverse=True)
print ("Top 3 words..")
pprint.pprint(sorted_candidates[:2])
```
